# Introduction to Matplotlib

Welcome to the introduction to Matplotlib. After watching this video, you'll be able to explain what Matplotlib is, why it was created, and describe its uses.

## Overview:
Matplotlib is one of the most widely used data visualization libraries in Python. It was created by John Hunter, an American neurobiologist, to address the limitations faced by his research team while working with proprietary software.

## Creation and Origin:
- **Creator:** John Hunter
- **Origin:** Developed during the analysis of electro CTO cartography and ECoG signals, replacing proprietary software.
- **Motivation:** Overcame the limitation of a single license by creating a Matlab-based version for collaborative use.

## Matplotlib's Architecture:
Matplotlib's architecture consists of three main layers:
1. **Backend Layer:**
   - Handles heavy lifting and is suitable for web applications, UI applications, and shared scripts.
   - Includes abstract interface classes like FigureCanvas, Renderer, and Event.


2. **Artist Layer:**
   - Responsible for creating visual elements in a Matplotlib figure.
   - Two types of artist objects: Primitive (e.g., Line2D, Rectangle) and Composite (e.g., Axis, Tick).
   


3. **Scripting Layer (Pyplot):**
   - Offers a lighter scripting interface for everyday tasks.
   - Matplotlib.Pyplot interface connects Canvas and Figure artist instances automatically.
   


## Scripting Example:
Generate a histogram of 10,000 random numbers using the Pyplot scripting layer:

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
X = np.random.randn(10000)

# Create a histogram with Pyplot
plt.hist(X, bins=100)

# Show the figure
plt.show()
```

Understanding the components and elements that make up a visual representation:
- Main components: Canvas, Axis, Title, Labels, Legend, Grid, Annotations.
- Components such as Title, Labels, Legend, Grid, and Annotations enhance the plot's readability and provide additional information.


````python
import matplotlib.pyplot as plt
%matplotlib inline
# %matplotlib inline is a magic command that allows plots to be embedded in the notebook itself and not open separately in another window
import numpy as np

# Generate random data
X = np.random.randn(10000)
print(X)
# Create a histogram with Pyplot
plt.hist(X, bins=100)
# Show the figure
plt.show()
````

## Basic Plotting with matplotlib

- Matplotlib is a well-established data visualization library that can be integrated into different environments such as Python scripts, Python and IPython shells, web application servers, and In Graphical User Interface toolkits. The Jupyter notebook is also one of them. It is an open-source web application that allows you to create and share documents containing live code, visualizations, and explanatory text.

- Jupyter has specialized support for Matplotlib, so if you start a Jupyter notebook, all you have to do is import Matplotlib, and you're ready to go. We will now learn how to create almost all visualization tools using the Scripting interface.

````python
plt.plot(1,5, 'x') #can also use 'o' for a circular point
plt.show()
````

One limitation of this backend is that you cannot modify a figure once it's rendered.
So after rendering the figure, there is no way for us to add, for example, a figure title or labels to its axis. We will need to generate a new plot and add a title and
the axis labels before calling the show function. 

````python
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plotting Example")
plt.show()
````

The point from the previous plot isn't visible anymore.

A backend that overcomes this limitation is the notebook backend. With the notebook backend in place, if a plt function is called,it applies them to the active figure if it exists. If a figure does not exist, it renders a new figure.So when we call the plt.plot function to plot a circular mark at position 5-5, the backend checks if an active figure exists. Since no active figure exists, it generates a figure and adds a circular mark to position 5-5. And what is beautiful about this backend is that now we can easily add a title or labels to the axis after the plot was rendered without the need of regenerating the figure. 

The above ```%matplotlib notebook``` command works on browser window, but not on VSCode. Use %matplotlib inline instead with a loop

````python
# Import the necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Enable the inline backend for static plots
%matplotlib inline

# Create a simple plot using the plt.plot function
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Check if an active figure exists; if not, create a new figure
if not plt.gcf().get_axes():
    plt.plot(x, y, label='Sine Wave')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Simple Sine Wave Plot')
    plt.legend()

# Update the plot by adding a vertical line after it has been rendered
plt.axvline(x=np.pi, color='red', linestyle='--', label='Vertical Line at π')
plt.legend()

# Display the plot
plt.show()
````

## Matplotlib - Pandas
Finally, another great aspect of Matplotlib is that Pandas also has a built-in implementation of it. Therefore, plotting in Pandas is as simple as calling the Plot function on a given Pandas series or Pandas data frame.

````python

import pandas as pd
data = {
    'Year': list(range(1980, 1996)),
    'India': [8880, 8670, 8147, 7338, 5704, 4211, 7150, 10189, 11522, 10343, 13844, 16700, 15110, 9930, 10015, 7895],
    'China': [5123, 6682, 3308, 1863, 1527, 1816, 1960, 2643, 2758, 3392, 4562, 6107, 5921, 5376, 4940, 4766]
}

india_china_df = pd.DataFrame(data)
india_china_df
````

For instance, if you have a data frame of the number of immigrants to Canada from India and China for the years 1980 to 1996 and you're interested in generating a line plot of this data, all you have to do is call the plot function on this data frame (India_China_df) and pass in kind='line'. There you have it, a line plot of the data in the data frame.

````python
india_china_df.plot(kind="line")
````

Plotting a histogram of the data is not any different. Suppose you want to plot a histogram of the India column in your data frame. In that case, all you have to do is call the plot function on that column and pass in kind='hist'. There you have it, a histogram of the number of Indian immigrants to Canada from 1980 to 1996.

````python
india_china_df["India"].plot(kind="hist")
````
